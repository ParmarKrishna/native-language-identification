## Base Variables

```python
MAIN_DIR_PATH = '/content/drive/Shareddrives/speech_communication'
MAIN_DB_PATH = '/content/drive/Shareddrives/speech_database'
MAIN_DATALOADER_ROOT = '/content/drive/Shareddrives/speech_dataloader_root'
```

```python
main_classes = ['Hindi','Kannada','Malayalam','Tamil','Telugu']
```

```python
os.chdir(MAIN_DB_PATH)
```

```python
folders = os.listdir(MAIN_DB_PATH)
if 'scripts' in folders:
  folders.remove('scripts')
if '.ipynb_checkpoints' in folders:
  folders.remove('.ipynb_checkpoints')
folders
```

```python
import pandas as pd
all_files = []
for folder in folders:
  os.chdir(f'{MAIN_DB_PATH}/{folder}/')
  os.chdir(folder)
  os.chdir('RECS')
  path = os.getcwd()
  # prPurple(f"Current PATH {path}")
  recordings = os.listdir()
  if 'find_speech_data_duration.py' in recordings:
    recordings.remove('find_speech_data_duration.py')
  count = len(recordings)
  for recording in recordings:
    if os.path.isdir(recording):
      os.chdir(recording)
      f = os.listdir()
      for files in f:
        filename=os.path.join(files)
        if folder=='Hindi':
          all_files.append([f'{path}/{recording}/{filename}',0])
        if folder=='Kannada':
          all_files.append([f'{path}/{recording}/{filename}',1])
        if folder=='Malayalam':
          all_files.append([f'{path}/{recording}/{filename}',2])
        if folder=='Tamil':
          all_files.append([f'{path}/{recording}/{filename}',3])
        if folder=='Telugu':
          all_files.append([f'{path}/{recording}/{filename}',4])
    count=count-1
    # prGreen(f"Completed with {recording}, remaining {count}")
    os.chdir(path)
  os.chdir(MAIN_DB_PATH)

  all_files
```

```python
print(all_files)
```
